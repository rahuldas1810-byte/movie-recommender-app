```python
# Merging both files based on title
movies = movies.merge(credits,on="title")
```

```python
# improve not used column
# genres
# movie_id
# keywords
# title
# overview
# cast
# crew
movies=movies[["movie_id","title","overview","genres","keywords","cast","crew"]]
```

```python
movies.dropna(inplace=True)
```

```python
#'[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]
# convert to ["Action","Adventure","Fantasy","SCiFi"]
```

```python
# string to list
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i["name"])
    return L
        
```

```python
movies["genres"]=movies["genres"].apply(convert)
```

```python
movies.sample()
```

```python
# do same for keywords
movies["keywords"]=movies["keywords"].apply(convert)
```

```python
movies.sample()
```

```python
# for cast take only top 3 cast
def convert3(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=3:
            
            L.append(i["name"])
            counter+=1
        else:
            break
    return L
        
```

```python
movies["cast"]=movies["cast"].apply(convert3)
```

```python
movies.sample()
```

```python
movies["crew"][0]
```

```python
# for crew only extract director name 
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i["job"]=="Director":
            L.append(i["name"])
            break
    return L
        
```

```python
movies["crew"]=movies["crew"].apply(fetch_director)
```

```python
movies.head()
```

```python
# overview is a string we have to convert it to list so that we merge 
movies["overview"]=movies["overview"].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
movies["genres"]=movies["genres"].apply(lambda x:[i.replace(" ","") for i in x])
movies["keywords"]=movies["keywords"].apply(lambda x:[i.replace(" ","") for i in x])
movies["cast"]=movies["cast"].apply(lambda x:[i.replace(" ","") for i in x])
movies["crew"]=movies["crew"].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.sample()
```

```python
movies["tags"] = movies["overview"] + movies["genres"] + movies["keywords"] + movies["cast"] + movies["crew"]
```

```python
movies.sample()
```

```python
new_df=movies[["movie_id","title","tags"]]
```

```python
new_df
```

```python
# covert tags list to string
new_df["tags"]=new_df["tags"].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
new_df["tags"][0]
```

```python
new_df["tags"]=new_df["tags"].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
# we have to find similarity between two movies tag so that we can recommend based on that
# to do this we have to apply text vectorization in the tag column
# apply bag of words vectorization
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words="english")
```

```python
vectors=cv.fit_transform(new_df["tags"]).toarray()
```

```python
# like convert same things word like ["loved","loving","love"]
# to ["love","love","love"]
%pip install nltk
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python

new_df["tags"]=new_df["tags"].apply(stem)
```

```python
# now find cosine distance to find similarity(angle between them)
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vectors)
 
```

```python
def recommend(movie):
    movie_index = new_df[new_df["title"] == movie].index[0]
    distances = similarity[movie_index]

    movies_list = sorted(
        list(enumerate(distances)),
        reverse=True,
        key=lambda x: x[1]
    )[1:6]

    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend("Avatar")
```
